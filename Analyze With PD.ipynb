```python
import pandas as pd
from IPython.display import display

INPUT_FILE = r"D:\Analysis\Plastic & Rubber - Part Family Wise.xlsx"

df = pd.read_excel(INPUT_FILE)
df.columns = df.columns.str.strip()

display(df.head())
```

```python
print(f"üìä Total Rows      : {len(df)}")
print(f"üìê Total Columns  : {df.shape[1]}")
print("\nüß© Column Names:")
display(pd.DataFrame(df.columns, columns=["Column Name"]))
```

```python
PART_COL = "PartNo"
PRICE_COL = "PO Price"

test_part = "KE220419"

display(
    df[df[PART_COL] == test_part][[PART_COL, PRICE_COL]]
)
```

```python
PART_FAMILY_COL = "Part Family"

distinct_count = df[PART_FAMILY_COL].nunique()
print(f"üß© Distinct Part Families: {distinct_count}")
```

```python
from IPython.display import display

part_families = (
    df[PART_FAMILY_COL]
    .dropna()
    .drop_duplicates()
    .sort_values()
    .reset_index(drop=True)
)

display(part_families.to_frame(name="Part Family"))
```

```python
part_family_counts = (
    df[PART_FAMILY_COL]
    .value_counts()
    .reset_index()
)

part_family_counts.columns = ["Part Family", "No of Records"]

display(part_family_counts)
```

```python
part_family_counts["% Share"] = (
    part_family_counts["No of Records"] / part_family_counts["No of Records"].sum() * 100
).round(2)

display(part_family_counts)
```

```python
PART_FAMILY_COL = "Part Family"

part_families = (
    df[PART_FAMILY_COL]
    .dropna()
    .drop_duplicates()
    .sort_values()
    .tolist()
)

print(f"Total Part Families: {len(part_families)}")
part_families[:10]   # preview first 10
```

```python
selected_family = part_families[0]   # or type manually
print("Selected Part Family:", selected_family)
```

```python
df_family = df[df[PART_FAMILY_COL] == selected_family]

print(f"Rows in '{selected_family}': {len(df_family)}")
display(df_family.head())
```

```python
PART_COL = "PartNo"

parts = df_family[PART_COL].dropna().unique()
print(f"Distinct Parts in '{selected_family}': {len(parts)}")

display(pd.DataFrame(parts, columns=["PartNo"]))
```

```python
selected_part = parts[0]
print("Selected Part:", selected_part)

PRICE_COL = "PO Price"

df_part = df_family[df_family[PART_COL] == selected_part]

display(df_part[[PART_COL, PRICE_COL]])
```

```python
df_part[PRICE_COL] = pd.to_numeric(df_part[PRICE_COL], errors="coerce")

print("Min PO Price:", df_part[PRICE_COL].min())
print("Max PO Price:", df_part[PRICE_COL].max())
```

```python
SUPPLIER_COL = "Vendor"

supplier_count = df_part[SUPPLIER_COL].nunique()
source_type = "Single Source" if supplier_count == 1 else "Multi Source"

print("Supplier Count:", supplier_count)
print("Source Type:", source_type)
```

```python
family_summary = (
    df_family
    .groupby(PART_COL)
    .agg(
        Min_PO_Price=("PO Price", "min"),
        Max_PO_Price=("PO Price", "max"),
        Supplier_Count=("Vendor", "nunique")
    )
    .reset_index()
)

display(family_summary)
```
